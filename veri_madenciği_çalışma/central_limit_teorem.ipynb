```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Popülasyon Tanımlama (örnek olarak exponential dağılım kullanalım)
np.random.seed(42)
population = np.random.exponential(scale=50, size=10000)

# 2. Popülasyon Özellikleri
pop_mean = np.mean(population)
pop_std = np.std(population)

# 3. Tek Seferlik Örnekleme
sample_30 = np.random.choice(population, size=30, replace=False)
sample_mean_30 = np.mean(sample_30)
sample_std_30 = np.std(sample_30)

# 4. Çoklu Örnekleme (100 kez)
sample_means_100 = [np.mean(np.random.choice(population, size=30, replace=False)) for _ in range(1000)]

sample_means_100_mean = np.mean(sample_means_100)
print(sample_means_100_mean)
```

```python
# 5. Görselleştirme
# Popülasyon Histogramı
plt.figure(figsize=(12, 6))
plt.hist(population, bins=50, alpha=0.7, label='Population', color='blue', density=True)
plt.axvline(pop_mean, color='red', linestyle='dashed', linewidth=2, label=f'Pop Mean: {pop_mean:.2f}')
plt.title('Population Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

# Örneklem Ortalama Dağılımı
plt.figure(figsize=(12, 6))
plt.hist(sample_means_100, bins=20, alpha=0.7, label='Sample Means (100)', color='green', density=True)
plt.axvline(sample_means_100_mean, color='red', linestyle='dashed', linewidth=2, label=f'Pop Mean: {sample_means_100_mean:.2f}')
plt.title('Distribution of Sample Means (n=30, repeated 100 times)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()
plt.show()

# Tek Seferlik Örneklem Özeti
(sample_mean_30, sample_std_30, pop_mean, pop_std)
```
